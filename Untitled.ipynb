```python
dos_cols = {'#':'E','E':'dosup(E)','(eV)':'dosdw(E)','dosup(E)':'Int'}
for i in range(len(dos_plots)):
    dos = dos_plots[i]
    dos = dos[['#', 'E', '(eV)', 'dosup(E)']]
    dos = dos.rename(dos_cols,axis='columns')
    dos_plots[i] = dos
```

```python
#dos_cu, dos1,dos2,dos3,dos4,dos5,dos6,dos7,dos8,dos9,dos_ni = dos_plots
dos_cu, dos1,dos2,dos3,dos4,dos5,dos6,dos9,dos_ni = dos_plots
```

```python
fig = plt.figure(figsize=(20,15))
for i in range(1,9):
    dos = dos_plots[i]
    fermi = fermis[i]
    dband = dbands[i]
    plt.ylim(-170,170)
    plt.xlim(-2.5,0.5)
    plt.plot(dos['E']-fermi.Energy[1],dos['dosup(E)'],lw = 2,color =comp_colors[i])
    plt.plot(dos['E']-fermi.Energy[1],dos['dosdw(E)']*-1,lw=2,color = comp_colors[i],label = names[i])
    plt.plot(fermi['Energy']-fermi.Energy[1],fermi['#']*-1,lw=2,color = 'black')
    plt.plot(dband['Energy']-fermi.Energy[1],dband['#']*-1,lw=2,color = comp_colors[i],ls='dashed')
    plt.legend(title='Compositions',title_fontsize=22,fontsize=22)
    plt.xlabel('E-EF(eV)',fontsize=30)
    plt.ylabel('DOS',fontsize=30)
    plt.xticks(fontsize=28)
    plt.yticks(fontsize=28)
plt.tight_layout()
plt.savefig('./all_dos.pdf',dpi=500)
```
